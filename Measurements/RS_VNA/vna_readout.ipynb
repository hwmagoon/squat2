# VNA Setup

[HM]: Note that there is a switch to enable averaging.  I'm not seeing a python function to control this feature.  So if you want to see the effects of your averaging, make sure to manually enable averaging on the device itself.

```python
## Daily measurement settings
warm_att = 0 #dB
## Define a savepath
today = datetime.date.today()
print(today)

expt_path = 'C:\_Data' + '\\' + "SQUAT_" + str(today)  + '\\' + "low_power"
print('Data will be stored in', expt_path)
```

```python
# Measurement Parameters
averages = 100 #3000
freq_start = 5e9 #Hz 
freq_stop  = 7e9 #7e9 #Hz
nb_points  = 10000 #max is 20000- dont do more! otherwise freq_sweep array is a lie, and plotting breaks
bandwidth  = 500 #Hz
power      = -60 #-50 #dBm
scattering_parameter = ('S12',) # needs to be tuple 
trace_name = ('trace1',) # needs to be tuple

## VNA setup
freq_sweep = np.linspace(freq_start, freq_stop, nb_points)
VNA.initialize_one_tone_spectroscopy(trace_name, scattering_parameter)
VNA.set_startfrequency(freq_start)
VNA.set_stopfrequency(freq_stop)
VNA.set_points(nb_points)
VNA.set_power(power)
VNA.set_measBW(bandwidth)
VNA.set_averages(averages)
VNA.set_averagestatus(status = 'on')
VNA.set_sweeps(averages)

## Sanity check power setting
power_at_device = power - warm_att - 90
print("Power on chip:", power_at_device, "dBm")
```

```python
VNA.initialize_one_tone_spectroscopy(trace_name, scattering_parameter)
```

# VNA Measurement

```python
while True:        
## Do measurement
        VNA.measure()

        ## Parse and store data
        [amps, phases] = VNA.get_traces(('trace1',))[0]
        data = {"series" : datetime.datetime.now().strftime('%Y%m%d_%H%M%S'),
                "amps": amps, 
                "phases" : phases, 
                "freqs": freq_sweep, 
                "vna_power": power, 
                "power_at_device" : power_at_device, 
                "bandwidth": bandwidth, 
                "averages" : averages,
                "nb_points": nb_points
                }

        fp1, fn1 = write_file(data, expt_path, filename="low_power_scans_to_avg")

        plot_amp(data, filepath=expt_path)
        plot_unwrapped_phase(data, filepath=expt_path)
```

```python
plot_amp(data, filepath=expt_path)
plot_unwrapped_phase(data, filepath=expt_path)
```

# Demonstrate reading data from file

```python
powers = [0, -10, -20]
avgs = [3, 15, 30]

filenames = []

for i in range(len(powers)):
    VNA.set_power(powers[i])
    VNA.set_averages(avgs[i])
    VNA.set_sweeps(avgs[i])
    VNA.measure()
    [amps, phases] = VNA.get_traces(('trace1',))[0]
    data = {"series" : datetime.datetime.now().strftime('%Y%m%d_%H%M%S'),
            "amps": amps, 
            "phases" : phases, 
            "freqs": freq_sweep, 
            "vna_power": powers[i], 
            "power_at_device" : powers[i] - warm_att - 90, 
            "bandwidth": bandwidth, 
            "averages" : avgs[i],
            "nb_points": nb_points
            }
    fp1, fn1 = write_file(data, expt_path, filename="zoomin5_5")
    filenames.append(fn1)
    plot_amp(data, filepath=expt_path)
    plot_unwrapped_phase(data, filepath=expt_path)
```

```python

for i in range(len(filenames)):
    fn = filenames[-(i+1)]
    print(fn)
    data = read_file(filename=(fn+".npz"), filepath=expt_path)
    plt.plot(data['freqs'], data['amps'], label=f"{data['power_at_device']} dBm")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude (dB)")
plt.legend()
plt.savefig(expt_path + '\\' + "power_scan_5_5.png")
plt.show()
```
